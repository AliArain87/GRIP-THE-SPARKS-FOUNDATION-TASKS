## **Library Section**
___

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import plotly.express as px
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

## **Load The Dataset**
___

```python
df = pd.read_csv('Iris.csv')
df.head()
```

## **Total Columns and Rows**
___

```python
rows, columns = df.shape
print('Number of rows: ', rows)
print('Number of columns: ', columns)
```

## **Data Preprocessing**
___

```python
df.info()
```

```python
df.describe()
```

```python
df.drop('Id', axis=1, inplace=True)
```

```python
df 
```

## **Splitting the dataset to into x and y**
___

```python
X = df.drop('Species', axis=1)
y = df[['Species']]
```

```python
X
```

```python
y
```

## **Training and Testing Phase**
___

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

## **Decision Tree Classifier**
___

```python
model = DecisionTreeClassifier()
```

## **Model Training**
___

```python
model.fit(x_train, y_train)
```

```python
print('Accuracy of Decision Tree classifier on training set: {:.2f}'.format(model.score(x_train, y_train)))
```

```python
print('Accuracy of Decision Tree classifier on testing set: {:.2f}'.format(model.score(x_test, y_test)))
```

## **Prediction**
___

```python
y_pred = model.predict(x_test)
y_pred
```

## **Save the model to disk**
___

```python
import pickle
filename = 'model.pkl'
pickle.dump(model, open(filename, 'wb'))
```

## **load the model from disk**
___

```python
loaded_model = pickle.load(open(filename, 'rb'))
```
